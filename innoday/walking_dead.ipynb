## Prepare data
Separate episode title from episode number:

```python
rawdata['title'] = rawdata.episode.apply(lambda x: x.split('-', 1)[1].strip())
rawdata.episode = rawdata.episode.apply(lambda x: x.split('-', 1)[0].strip())
```

```python
rawdata
```

Split episode number into the 'season' and 'episode' part:

```python
rawdata['season'] = rawdata.episode.apply(lambda x: int(x.split('x', 1)[0]))
rawdata.episode = rawdata.episode.apply(lambda x: int(x.split('x', 1)[1]))
```

```python
rawdata
```

Check if our dataframe uses the correct datatypes:

By 'exploding' the text attribute we instruct pandas to create a separate row for each line in our dataset:

```python
df = rawdata.explode('text')
```

```python
df
```

Filter out rows where a character is mentioned at the beginning of the line:

```python
filtered = df[df.text.str.match('^[a-zA-Z]{1,15}:.*$')]
```

Count the number of rows, grouped by season and episode:

```python
stats = filtered.drop(columns='title').groupby(['season', 'episode']).count()
```

Define a function for extracting the speaking character from a single line:

```python
import re
def extract_name(line):
  m = re.search('^([a-zA-Z]{1,15}):.*$', line)
  if(m):
    return m.group(1)
  else:
    return '' # no character is mentioned
```

Total number of mentions per character:

```python
df.text.apply(extract_name).value_counts()
```

Character count for entire season 1:

```python
df[df.season==1].text.apply(extract_name).value_counts()
```

Character count for episode 1 of season 1

```python
df[(df.season==1) & (df.episode==1)].text.apply(extract_name).value_counts()
```

Character count for episode 2 of season 1

```python
df[(df.season==1) & (df.episode==2)].text.apply(extract_name).value_counts()
```

Character count for episode 3 of season 1

```python
df[(df.season==1) & (df.episode==3)].text.apply(extract_name).value_counts()
```

Character count for episode 4 of season 1

Split into train and test set:

```python
!pip3 install sklearn
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

Define our machine learning pipeline:

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.ensemble import GradientBoostingRegressor

text_clf = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('clf', GradientBoostingRegressor(random_state=420)),
])
```

Fit the model:

```python
text_clf.fit(X_train,y_train)
```

Evaluate the model on the test set:

```python
text_clf.score(X_test,y_test)
```

Prepare hyperparameter optimization:

```python
from sklearn.model_selection import RandomizedSearchCV

grid = {
    'vect__ngram_range': [(1, 1), (2, 1)],
    'vect__stop_words': [None, 'english'],
    'clf__learning_rate':[0.15,0.1,0.05,0.01,0.005,0.001], 
    'clf__n_estimators':[100,250,500,750,1000,1250,1500,1750], 
    'clf__max_depth':[2,3,4,5,6,7]
}
```

Run optimization:

```python
search = RandomizedSearchCV(text_clf, grid, n_iter=100, n_jobs=-1, random_state=42, verbose=2)
search.fit(X_train, y_train)
```

Evaluate the optimized model:

```python
search.score(X_test, y_test)
```

Get model predictions for each episode:

```python
rawdata['predicted_deaths'] = rawdata.text.apply(lambda x: search.predict([str(x)])[0])
```

```python
rawdata
```

Plot the predicted and actual death count for each episode:

```python
import seaborn as sns

# define colors of the two lines
colors = [
          (0,0,1.0,0.8),
          (1.0,0.6,0,0.8)
]

sns.relplot(x='episode', 
            col='season', 
            kind='line', 
            y='value', 
            hue='variable',
            palette = colors, 
            linewidth = 3,
            data=pd.melt(rawdata, ['season', 'episode'], # transform data to long format for seaborn
                         value_vars=['deaths', 'predicted_deaths']), 
            col_wrap=4);
```
